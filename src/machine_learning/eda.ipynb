```python


from preprocess import *
from outlier_detection import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import phik
from phik import resources, report


preprocessor = Preprocess(override=True)
```

```python
def eda(preprocessor: Preprocess):
    
    # Load the fire data
    df = preprocessor.get_fire_data()
    
    for province in df['ProvinceID'].unique():
        directory = f"./images/phik"
        os.makedirs(directory, exist_ok=True)
        
        fire_data = df[df['ProvinceID'] == province]
        print(fire_data.columns)
        fire_data = fire_data[['AverageTemperature', 'HectaresBurnt', 'MaxRelativeHumidity', 'MaxWindspeedGust', 'FireLatitude', 'FireLongitude']]
        
        phik = fire_data.phik_matrix()
        sns.heatmap(phik, cmap='coolwarm', annot=True)
        plt.title(f"Phik Matrix for {preprocessor.province_mapping[province]}")
        plt.tight_layout()
        plt.savefig(f'{directory}/{preprocessor.province_mapping[province]}.png')
        plt.close()
        
        report.correlation_report(fire_data, pdf_file_name=f'{directory}/{preprocessor.province_mapping[province]}_correlations.pdf')
        
        

eda(preprocessor)
```
